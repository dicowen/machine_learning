```python
# features (1 sim, 0 não)
# pelo longo? 
# perna curta?
# faz auau?
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

# 1 => porco, 0 => cachorro
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
treino_y = [1, 1, 1, 0, 0, 0]
```

```python
from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(treino_x, treino_y)
```

```python
animal_misterioso = [1, 1, 1]
model.predict([animal_misterioso])
```

```python
misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]

testes_x = [misterio1, misterio2, misterio3]
testes_y = [0, 1, 1]

previsoes = model.predict(testes_x)
```

```python
corretos = (previsoes == testes_y).sum()
total = len(testes_x)
taxa_de_acerto = (corretos / total) * 100
print('A taxa de acerto foi: {}'.format(taxa_de_acerto.round(2)))
```

```python
from sklearn.metrics import accuracy_score

taxa_de_acerto = accuracy_score(testes_y, previsoes)
print('Taxa de acerto: {}'.format((taxa_de_acerto*100).round(2)))
```
